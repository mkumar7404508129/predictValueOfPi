# Linear random value generators

```python
import numpy as np
def uniform_good(size=1,seed=123456789, mul=16807,mod=(2**31)-1):
    u=np.zeros(size);
    x=(seed*mul+1)%mod
    u[0]=x/mod
    for i in range(1,size):
        x=(x*mul+1)%mod;
       
        u[i]=x/mod;
    return u;
```

```python
size=100
x=uniform_good(size=size)
print(x)
```

```python
y=[x/100 for x in range(0,100,10)]
print(y)
plt.hist(x,bins=y,ec='red')
plt.xticks(y)
plt.show()
```

```python

def randomBetween(low=0,heigh=1,size=1,seed=123456789):
 
    return low+(heigh-low)*uniform_good(seed=seed,size=size)
```

```python
x=randomBetween(low=10,heigh=20,size=1000)
print(x)
```

```python
y=[x for x in range(10,21)]
print(y)
plt.hist(x,bins=y,ec='red')
plt.xticks(y)
plt.show()
```

# Making seed using system clock()

```python
def random(low=0,heigh=1,mul=1234567890,mod=(2**30)-1):
    
    t=time.perf_counter()
    seed=int(10**9*(t-int(t)))
    x=(seed*mul+1)%mod
    x=x/mod
    value=low+(heigh-low)*x
    return value

```

```python
random(-1,1)
```

```python
# Printing random integer  value in range 0 and 100
data=[int(random()*100) for x in range(0,1000)]
print(data)
```

```python
y=[x for x in range(0,100,10)]
print(y)
plt.hist(data,y,ec="red")
plt.xticks(y)
plt.show()
```

# Pi value estimation using Above random function

pi value can be estimated by "Monte Carlo methods" using formula
pi=4*(total number of random value inside circle/total number of random value)  

```python
n=100000
```

```python
points=[[random(low=-1,heigh=1),random(low=-1,heigh=1)] for x in range(0,n)]
points
```

```python
points_inside_circle=[[x,y] for x,y in points if x**2 + y**2 <=1]
points_inside_circle
```

```python
4*(len(points_inside_circle)/len(points))
```

```python
# using random seed as system clock
def run_expriment(n=100):
    points=[[random(low=-1,heigh=1),random(low=-1,heigh=1)] for x in range(0,n)]
    points_inside_circle=[[x,y] for x,y in points if x**2 + y**2 <=1]
    return 4*(len(points_inside_circle)/len(points))
    
```

```python
result=[]
avearge=[]
for _ in range(1000):
    result.append(run_expriment())
    avg_pi=sum(result)/len(result)
    print(avg_pi)
    avearge.append(avg_pi)
```
